# What is Support Vector Machine?

An SVM model is a representation of the examples as points in space. The mapped points separate the categories by dividing them  by a clear gap as wide as possible.
SVM can efficiently perform linear as well as non-linear classification by implictly mapping the inputs into high dimensional feature spaces.


# What does SVM do?

The dataset in the training examples are categorised. Each point belongs to one or the other category. The SVM training algorithm builds a model to classify the new dataponints thereby, making it a non-probabilistic binary linear classifier.

```python
# importing scikit learn with make_blobs 
from sklearn.datasets.samples_generator import make_blobs 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
  
# creating datasets X containing n_samples 
# Y containing two classes 
X, Y = make_blobs(n_samples=500, centers=2, 
                  random_state=0, cluster_std=0.40) 
  
# plotting scatters  
plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='spring'); 
plt.show()  
```

```python
# creating line space between -1 to 3.5  
xfit = np.linspace(-1, 3.5) 
  
# plotting scatter 
plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='spring') 
  
# plot a line between the different sets of data 
for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]: 
    yfit = m * xfit + b 
    plt.plot(xfit, yfit, '-k') 
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',  
    color='#AAAAAA', alpha=0.4) 
  
plt.xlim(-1, 3.5); 
plt.show() 
```
