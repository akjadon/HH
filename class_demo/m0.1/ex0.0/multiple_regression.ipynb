## Demo 1 - Simple Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df['MEDV'] = boston.target
boston_df.head()
```

## Simple Linear Regression

```python
# Use only one feature RM - Average no of rooms
X1 = boston_df[['RM']]
y = boston_df.MEDV

from sklearn.linear_model import LinearRegression
slm1 = LinearRegression()
slm1.fit(X1, y) # fit the model
print("Coefficient:", slm1.coef_,"Intercept:", slm1.intercept_)
```

```python
predicted_1 = slm1.predict(X1) # Predict prices
print("Predicted values: ", predicted_1[:4])
print("Actual values:\n", boston_df[['RM','MEDV']][:4])
```
